```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams

#%matplotlib inline
```

```python
working_dir = "../../data/working/"
input_dir = "../../data/input/"

responses = pd.read_csv(input_dir + "responses.csv")
```

```python
#plt.figure(figsize=(9,7))
rcParams['figure.figsize'] = 11.7,8.27
sns.set(font_scale=1.5) 
matrix = responses.corr().round(2)
sns.heatmap(matrix, annot=True)
```

```python
sns.set_theme(style="darkgrid")
sns.set(font_scale=1.5) 
sns.countplot(x="Category", data=responses)
```

```python
sns.catplot(x="pId", hue="Category", kind="count",
            data=responses, height=9.27, aspect=15.7/9.27)
```

```python
sns.catplot(x="docId", hue="Engagement", kind="count",
            data=responses, height=9.27, aspect=17.7/9.27)
```

```python
sns.catplot(x="Category", y="Engagement", hue="pId", kind="bar", 
            data=responses, height=9.27, aspect=17.7/9.27)
```

```python
sns.relplot(x="Arousal", y="Engagement",hue= "Category",col="Category", data=responses)
```

```python
sns.relplot(x="Arousal", y="Engagement",hue= "Category",
            col="pId",col_wrap=4,kind='scatter', data=responses)
```
