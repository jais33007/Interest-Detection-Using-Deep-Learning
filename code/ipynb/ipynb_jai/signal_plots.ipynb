```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os
import neurokit2 as nk
%matplotlib inline
```

```python
#participant 2 and document 7 : high interest
main_dir = "/Users/jai/Documents/IEEE-Interest/interest/analysis/data/output/"
dir = main_dir + "E4_mapped/"

eda_df = pd.read_csv(dir + "p06/p06_d19_EDA.csv", index_col=0)
temp_df = pd.read_csv(dir + "p06/p06_d19_TEMP.csv", index_col=0)
bvp_df = pd.read_csv(dir + "p06/p06_d19_BVP.csv", index_col=0)
hr_df = pd.read_csv(dir + "p06/p06_d19_HR.csv", index_col=0)
```

```python
eda_x = eda_df.timestamp.values
temp_x = temp_df.timestamp.values
bvp_x = bvp_df.timestamp.values
hr_x = hr_df.timestamp.values

eda_y = eda_df.data.values
temp_y = temp_df.data.values
bvp_y = bvp_df.data.values
hr_y = hr_df.data.values
```

```python
eda_df1 = pd.read_csv(dir + "p06/p06_d18_EDA.csv", index_col=0)
temp_df1 = pd.read_csv(dir + "p06/p06_d18_TEMP.csv", index_col=0)
bvp_df1 = pd.read_csv(dir + "p06/p06_d18_BVP.csv", index_col=0)
hr_df1 = pd.read_csv(dir + "p06/p06_d18_HR.csv", index_col=0)
```

```python
eda_x1 = eda_df1.timestamp.values
temp_x1 = temp_df1.timestamp.values
bvp_x1 = bvp_df1.timestamp.values
hr_x1 = hr_df1.timestamp.values

eda_y1 = eda_df1.data.values
temp_y1 = temp_df1.data.values
bvp_y1 = bvp_df1.data.values
hr_y1 = hr_df1.data.values
```

```python
fig, axes = plt.subplots(4, 2, figsize=(50,25))
sns.set(font_scale = 1.8)

#fig.suptitle('Raw signals from the Empatica E4 based on a single participant response', fontsize=22)

sns.lineplot(ax=axes[0, 0], data=eda_df, x=eda_x, y=eda_y, color='green')
axes[0,0].set_ylabel('EDA',fontsize=24)
axes[0,0].set_xlabel('Interest="low"(1)',fontsize=24)
axes[0,0].xaxis.set_label_position('top')

sns.lineplot(ax=axes[0, 1], data=eda_df1, x=eda_x1, y=eda_y1, color='green')
axes[0,1].set_xlabel('Interest="high"(4)',fontsize=24)

axes[0,1].xaxis.set_label_position('top')
sns.lineplot(ax=axes[1, 0], data=bvp_df, x=bvp_x, y=bvp_y, color='orange')
axes[1,0].set_ylabel('BVP',fontsize=24)


sns.lineplot(ax=axes[1, 1], data=bvp_df1, x=bvp_x1, y=bvp_y1, color='orange')
sns.lineplot(ax=axes[2, 0], data=temp_df, x=temp_x, y=temp_y)
axes[2,0].set_ylabel('Temperature',fontsize=24)

sns.lineplot(ax=axes[2, 1], data=temp_df1, x=temp_x1, y=temp_y1)
sns.lineplot(ax=axes[3, 0], data=hr_df, x=hr_x, y=hr_y, color='red')
axes[3,0].set_ylabel('Heart Rate',fontsize=24)
axes[3,0].set_xlabel('Time',fontsize=24)
axes[3,1].set_xlabel('Time',fontsize=24)
sns.lineplot(ax=axes[3, 1], data=hr_df1, x=hr_x1, y=hr_y1,color='red')
plt.savefig("/Users/jai/Documents/IEEE-Interest/plots/E4_raw_ratings.pdf", bbox_inches='tight')
```
